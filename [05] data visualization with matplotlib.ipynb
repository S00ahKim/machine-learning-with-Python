# data visualize

1. 라인 플랏

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*5
print(x,y)
plt.plot(x,y)
plt.show()
```

라인플랏 - 그래프 이름과 축 라벨 추가

```python
plt.title('my gragh')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
plt.show()
```

선 대신 점으로 데이터 표현

```python
plt.plot(x, y, 'go') #green-o
plt.show()
```

2. 서브플랏으로 멀티플 그래프

```python
plt.figure() #캔버스 생성

plt.subplot(2, 1, 1) #두개의 그래프 중에 첫번째
x1 = np.array([1,2,3,4])
y1 = x1 * 5
plt.plot(x1, y1, 'ro')

plt.subplot(2, 1, 2)
x2 = np.array([3,4,5,6])
y2 = x2 * 3
plt.plot(x2, y2)

plt.show()
```

```python
# 그리드

from matplotlib import pyplot as plt
import numpy as np

x = np.array([1,2,3,4])
y = x*5

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.text(3, 14, 'Here is (3,4).')

plt.grid(True) #격자가 그려지게 된다.
plt.plot(x,y,'ro')

plt.show()
```

```python
# 산포도
from matplotlib import pyplot as plt
import matplotlib 
    # 내부 요소만 import할 게 아니라 전부 해줘야 
    # 아래에서 matplotlib.부분을 쓸 수 있다.
import numpy as np
from matplotlib import font_manager, rc

# 폰트에 따라 음수일 경우 음수 - 표기가 잘 되지 않는 경우가 있어 폰트 적용
font_loc = "./unicode.ttf"
font_name = font_manager.FontProperties(fname=font_loc).get_name()
matplotlib.rc('font', family=font_name)

#정규분포를 가지는 1차원 어레이 실수 1000개 만들기. 
# np.random.randn(n, m)의 경우 n*m개의 2차원 실수 생성
x = np.random.randn(1000)
y1 = np.random.randn(len(x))
y2 = np.exp(x)

plt.subplot(1,2,1)
plt.scatter(x, y1, s=1, alpha=0.5)
#s는 point의 사이즈를 의미한다. 알파는 투명도.

plt.xlim(-4.0, 4.0) #x축, y축의 범위
plt.ylim(-5, 30)
plt.grid()

plt.subplot(1,2,2)
plt.scatter(x,y2, s=1, alpha=0.5)
plt.xlim(-4.0, 4.0)
plt.ylim(-5, 30)
plt.grid()

plt.show()
```
