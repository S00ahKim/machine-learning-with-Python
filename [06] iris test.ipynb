```python
import pandas as pd

mr = pd.read_csv('mushroom.csv')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split

df=pd.DataFrame(mr.iloc[:,0])#한 칼럼만 선택하면 시리즈 되어서 다시 데이터프레임화

#두번째 칼럼부터 마지막까지 원핫인코딩, 라벨 붙임
df = df.join(pd.get_dummies(mr.iloc[:, 1:]))
print(df)
```

```python
data = df.iloc[:, 1:]
label = df.iloc[:, 0]

data_train, data_test, label_train, label_test = train_test_split(data, label, test_size=0.9)

clf = RandomForestClassifier(n_estimators=1, criterion="entropy") #랜포 트리갯수설정
clf.fit(data_train, label_train)

predict = clf.predict(data_test)

result=pd.DataFrame({'label':label_test, 'pre':predict})
print(result[0:10])

ac_score = metrics.accuracy_score(label_test, predict)
print("정답률 = ", ac_score)
```
